```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
premise_file = "../Resources/LicensePremise.csv"
premise_data_df = pd.read_csv(premise_file)
premise_data_df.head(2)
```

```python
county_file = "../Resources/CountyLicenseCount.csv"
county_data_df = pd.read_csv(county_file)
county_data_df.head(2)
```

### Transform premise DataFrame

```python
new_premise_df = premise_data_df[["Premises Name", "County ID Code"]]
new_premise_df.columns = ['premise_name', 'county_id']
new_premise_df.head()
new_premise_df['id']=new_premise_df.index
new_premise_df.head()

```

### Transform county DataFrame

```python
new_county_df = county_data_df[["County Name (Licensee)", "County ID Code"]]
new_county_df.columns = ['county_name', 'county_id']
new_county_df['id'] = new_county_df.index
new_county_df.head(2)
```

### Create database connection

```python
rds_connection_string = "root:12345678@127.0.0.1/customer_db"
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
new_premise_df.to_sql(name="premise", con=engine, if_exists="append", index=False)
```

```python
new_county_df.to_sql(name="county", con=engine, if_exists="append", index=False)
```
